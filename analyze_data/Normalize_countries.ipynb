```python
# input data

import pandas as pd
import re

data = []
with open('/Users/pengqiushi/Code/country.csv', 'r') as file:
    for line in file:
        data.append(re.split('[,]', line.strip(), maxsplit = 1))

raw_countries = pd.DataFrame(data)
raw_countries.columns = ['pmc_ID', 'country']
raw_countries = raw_countries.reindex(columns = raw_countries.columns.tolist() + ['converted_country_1', 'converted_country_2'])

raw_countries
```

```python
from flashgeotext.geotext import GeoText
geotext = GeoText()

for idx in range(len(raw_countries.loc[:, 'country'])):
    try:
        name = list(geotext.extract(input_text = raw_countries.loc[idx, 'country'])['countries'].values())[-1]['found_as'][0]
        raw_countries.converted_country_1[idx] = name
    except:
        pass
```

```python
import country_converter as coco

for idx in range(len(raw_countries.loc[:, 'country'])):
    try:
        standard_name = coco.convert(names = raw_countries.loc[idx, 'converted_country_1'], to = 'name_short')
    except:
        standard_name = 'N/A'
    raw_countries.converted_country_2[idx] = standard_name
```

```python
for index, row in raw_countries.iterrows():
    if row['converted_country_1'] == 'México':
        row['converted_country_2'] = 'Mexico'
    elif row['converted_country_1'] == 'UAE':
        row['converted_country_2'] = 'United Arab Emirates'
    elif row['converted_country_1'] == 'PRC':
        row['converted_country_2'] = 'China'
    elif row['converted_country_1'] in ['Nederland', 'Holland']:
        row['converted_country_2'] = 'Netherlands'
    elif row['converted_country_1'] == 'Perú':
        row['converted_country_2'] = 'Peru'
    elif row['converted_country_1'] in ['Saxony', 'Deutschland']:
        row['converted_country_2'] = 'Germany'
    elif row['converted_country_1'] in ['Sarawak', 'Sabah']:
        row['converted_country_2'] = 'Malaysia'
    elif row['converted_country_1'] == 'Slovenija':
        row['converted_country_2'] = 'Slovenia'
    elif row['converted_country_1'] =='España':
        row['converted_country_2'] = 'Spain'
    elif row['converted_country_1'] == 'Norge':
        row['converted_country_2'] = 'Norway'
    elif row['converted_country_1'] == 'Aotearoa':
        row['converted_country_2'] = 'New Zealand'
    else:
        pass
```

```python
formal_countries = raw_countries.dropna(axis = 0,how = 'any').loc[:, ['pmc_ID', 'converted_country_2']]
```

```python
formal_countries.rename(columns={'converted_country_2':'country'}, inplace = True)
```

```python
formal_countries.to_csv('./formal_countries.csv', index = False)
```
